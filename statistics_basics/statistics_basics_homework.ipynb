столбец respiratory rate. Максимальное значение 96, а по описанию 8-10 (хотя в описании написано, что 
полезность столбца сомнительна, из-за больших отклонений)
Все остальные столбцы в содержат значения в допустимой норме, исходя из описания.


  1:  surgery?
          1 = Yes, it had surgery
          2 = It was treated without surgery

  2:  Age
          1 = Adult horse
          2 = Young (< 6 months)
  4:  rectal temperature
          - linear
          - in degrees celsius.
          - An elevated temp may occur due to infection.
          - temperature may be reduced when the animal is in late shock
          - normal temp is 37.8
          - this parameter will usually change as the problem progresses
               eg. may start out normal, then become elevated because of
                   the lesion, passing back through the normal range as the
                   horse goes into shock
  5:  pulse
          - linear
          - the heart rate in beats per minute
          - is a reflection of the heart condition: 30 -40 is normal for adults
          - rare to have a lower than normal rate although athletic horses
            may have a rate of 20-25
          - animals with painful lesions or suffering from circulatory shock
            may have an elevated heart rate
  6:  respiratory rate
          - linear
          - normal rate is 8 to 10
          - usefulness is doubtful due to the great fluctuations

  7:  temperature of extremities
          - a subjective indication of peripheral circulation
          - possible values:
               1 = Normal
               2 = Warm
               3 = Cool
               4 = Cold
          - cool to cold extremities indicate possible shock
          - hot extremities should correlate with an elevated rectal temp.
   11: pain - a subjective judgement of the horse's pain level
          - possible values:
               1 = alert, no pain
               2 = depressed
               3 = intermittent mild pain
               4 = intermittent severe pain
               5 = continuous severe pain
          - should NOT be treated as a ordered or discrete variable!
          - In general, the more painful, the more likely it is to require
            surgery
          - prior treatment of pain may mask the pain level to some extent
   23: outcome
          - what eventually happened to the horse?
          - possible values:
               1 = lived
               2 = died
               3 = was euthanized

### принимаем решение выбросы по столбцу Age со значением '9' заменить на 1, 
т.к. это единственное значение,которое уже присутствует в столбце и соответствует лошади старше 6 месяцев

```python
# Посчитаем межквартильные размахи для определения выбросов по столбцу 'pulse'
q1 = horses_df['pulse'].quantile(0.25)
q3 = horses_df['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horses_df[horses_df['pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
remove_outliers
```

```python
#Что это за выбросы по столбцу 'pulse'?
pd.concat([horses_df, remove_outliers]).drop_duplicates(keep=False)
```

Кроме столбца outcome, нет ярко выраженных мод внутри столбцов
Будем искать взаимосвязь между различными столбцами- результат замены то этого только улучшится ;)

```python
# Гипотеза- проверим взаимосвязь между столбцами pain и outcome 
print(horses_df.groupby(['pain'])['outcome'].median())
```

можно сделать вывод, что при значениях pain(боль)-  4 (периодическая сильная боль) и 5(постоянная сильная боль), медианное значение столбца outcome равно 2 (смерть).
Поэтому недостающие значения в столбце outcome заполняеем исходя из значения столбца pain

```python
#Сначала заполним недостающими значениями столбец outcome (у него самый маленький процент пропусков)
#найдем это значение 
horses_df.loc[((horses_df['pain'] == 4) | (horses_df['pain'] == 5)) & horses_df['outcome'].isna()]
```

```python
#Все действия будем производить с новым датафреймом, для быстрого отката в случае неправильного применения фильтра
#Заполним недостающие значениями столбца outcome
fill_horses_df = horses_df
fill_horses_df['outcome'].fillna(fill_horses_df.groupby(['pain'])['outcome'].transform('median'), inplace=True)
fill_horses_df.describe()
```

```python
# Гипотеза- проверим взаимосвязь между столбцами surgery? и pain (тем более это описанов документации)
print(fill_horses_df.groupby(['surgery?'])['pain'].median())
```

Взаимосвязь есть- операция была (1) при высоком медианном значении боли(3), и не было(2) при низком значении боли(2)

```python
#Заполним недостающие значениями столбца pain исходя из медианных значений столбца surgery
fill_horses_df['pain'].fillna(fill_horses_df.groupby(['surgery?'])['pain'].transform('median'), inplace=True)
fill_horses_df.describe()
```

```python
# В подтверждение гипотезы роверим обратную взаимосвязь между столбцами pain(боль) и surgery(хтрургическое вмешательство)
print(horses_df.groupby(['pain'])['surgery?'].median())
```

взаимосвязь есть - что на уровнях боли 3 и выше всегда было хирургическое вмешательство. Восстановим отсутствующее значение в столбце 'surgery?' 

```python
fill_horses_df['surgery?'].fillna(fill_horses_df.groupby(['pain'])['surgery?'].transform('median'), inplace=True)
fill_horses_df.describe()
```
