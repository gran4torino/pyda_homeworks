```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

    
#Обработка ввода комманды 
def input_commands():
    commands_for_input_n_doc = ['p','s','ad','d','m']
    commands_for_input_n_shelf = ['ads','ds','m']
    commands_others = ['l','q']
    commands_all =commands_for_input_n_doc+commands_for_input_n_shelf+commands_others
    commands_name = input('Введите команду:')
    #проверка комманд на правильность
    if commands_name not in commands_all:
        print('Нет такой команды! Введите комманду из списка: p, s, ad, d, m, ads, ds, m, l, q')
    if commands_name in commands_for_input_n_doc:  
        document_num = input('Введите номер документа:')
    if commands_name in commands_for_input_n_shelf:
        shelf_number = input('Введите номер полки:')
    #Поиск владельца документа по его номеру
    if commands_name == 'p':
        owner_search_by_id(document_num)
    #По номеру документа узнать на какой полке он хранится
    if commands_name == 's':
        shelf_search_by_id(document_num,False)
    #увидеть полную информацию по всем документам
    if commands_name == 'l':
        full_documents_info() 
    #добавляем новую полку
    if commands_name == 'ads':
        add_new_folder(shelf_number) 
    #Удаляем существующую полку,если она непустая
    if commands_name == 'ds':
        remove_shelf(shelf_number) 
    #Добавляем новый документ
    if commands_name == 'ad':
        doc_type = input('Введите тип документа:')
        name = input ('Введите владельца документа:')
        shelf_number = input('Введите полку для хранения:')
        add_new_doc(doc_type,document_num,name,shelf_number)
     #Удаляем документ
    if commands_name == 'd':
        delete_doc(document_num)
    #Перемещаем  документ с полки на полку
    if commands_name == 'm':
        move_doc_shelf(document_num,shelf_number)   
    result = commands_name
    return result


def owner_search_by_id(document_num,only_number=False):
    '''
    'p'- комманда вызова
    Поиск владельца документа по его номеру
    only_number (True, False) - 
    если True - при отработке функции передавать только есть владелец или нет (True or False), без сопроводительного текста
    если False (по умолчанию)- то возращает с текстом "Владелец документа: ФИО"
    '''
    
    result = False
    for n in documents:
        if document_num in list(n.values()):
            if only_number:
                result = True
            else:    
                print('Владелец документа: ', n['name'])
                result = True
    if result:
        None
    else:
        if only_number:
            result = False
        else:
            print('Документ не найден в базе.')
        
        
        #print(result)
    return result

def shelf_search_by_id(document_num,only_number=True):
    '''
    's'- комманда вызова
    По номеру документа узнать на какой полке он хранится
    only_number (True, False) - 
    если True(по умолчанию) - при отработке функции передавать только номер полки, без сопроводительного текста
    если False- то возращает с текстом "Документ хранится на полке:"

    '''
    found = False
    for n in list(directories.values()):
         if document_num in n:
            #номер полки = возвращаем значения ключа в списке[находим индекс с помощью .index(n), где n это список,
            #который содержит номер искомого документа
            result = list(directories.keys())[list(directories.values()).index(n)]
            if only_number:
                result
            else:
                print('Документ хранится на полке:', result)
            found = True
    if found:
        None
    else:
        if only_number:
            result = False
        else:
            print('Документ не найден в базе.')
            result = False
    return result

def full_documents_info():
    '''
    'l'- комманда вызова
    Смотрим полную информацию по всем документам
    '''
    for n in documents:
        print ('№:',n['number'],', тип:', n['type'],', владелец:', n['name'],', полка хранения: ',shelf_search_by_id(n['number'])) 
    return 
def current_shelfes_list():
    '''
    Подсчитываем и печатаем текущий перечень полок
    '''
    z = 0
    for i in  directories.keys():
        z += 1
        if z != len(directories.keys()):
            print(i,end = ', ')
        else:
            print(i,'.')
    return

def add_new_folder(shelf_number):
    '''
    'ads'- комманда вызова
    Добавляем новую полку
    shelf_number- номер полки
    '''
    if shelf_number not in directories.keys():
        message = ('Полка добавлена.')
    else:
        message = ('Такая полка уже существует.')
    directories.setdefault(shelf_number,[])
    print(message, "Текущий перечень полок: ",end='')
    current_shelfes_list()
    return

def remove_shelf(shelf_number):
    '''
    'ds'- комманда вызова
    Удаляем полку, только если она пустая 
    shelf_number- номер полки
    '''
    if shelf_number not in directories.keys():
        message = ('Такой полки не существует. ')
    elif shelf_number in directories.keys() and len(list(directories.get(shelf_number)))== 0:
        del directories[shelf_number]
        message = ('Полка удалена.')
    else:
        message = ('На полке есть документы, удалите их перед удалением полки. ')
    print(message,end='')
    print("Текущий перечень полок: ",end='')
    current_shelfes_list()
    return
def add_new_doc(doc_type,document_num,name,shelf_number):
    '''
    'ad'- комманда вызова
    Добавляем новый документ
    '''
    if shelf_number not in directories.keys():
        print('Такой полки не существует. Добавьте полку командой ads.')
    else:
        documents.append({'type':doc_type,'number':document_num,'name':name})
        add_doc_on_shelfes(document_num,shelf_number)
        print('Документ добавлен. Текущий список документов:')
    full_documents_info()
    return
def add_doc_on_shelfes(document_num,shelf_number):
    '''
    Добавляем документ по его номеру на полку 
    '''
    result = directories[shelf_number].append(document_num)
    return result
def delete_doc_on_shelfes(document_num):
    '''
    Удаляем документ по его номеру c полки 
    '''
    result = directories[str(shelf_search_by_id(document_num))].remove(document_num)
    return result
def delete_doc(document_num):
    '''
     'd'- комманда вызова
     Удаляем документ по его номеру c перечня документов и полки 
    '''
    if shelf_search_by_id(document_num):
        directories[str(shelf_search_by_id(document_num))].remove(document_num)
        z = 0
        for element in documents:
            if document_num == documents[z]['number']:
                del documents[z]
            z += 1
        print('Документ удален.') 
    else:
        print('Документ не найден в базе.')
    full_documents_info()
    return
def move_doc_shelf(document_num,shelf_number):
    '''
    'm'- комманда вызова
    Перемещаем  документ с полки на полку
    '''
    if shelf_number not in directories.keys():
        print('Такой полки не существует. ')
        print("Текущий перечень полок: ",end='')
        current_shelfes_list()
    #elif True:
    elif owner_search_by_id(document_num,True):
        delete_doc_on_shelfes(document_num)
        add_doc_on_shelfes(document_num,shelf_number)
        print('Документ перемещен.')
        full_documents_info()
    else:
        print('Документ не найден в базе.')
        print('Текущий список документов:')
        full_documents_info()
    return
    
while input_commands() !='q':
    None
print('Остановка программы!')
```
